```python
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
import os

def parse_pdf_with_docling(pdf_path: str):
    file_name = os.path.basename(pdf_path)

    # Configure PDF pipeline with detailed options
    pdf_options = PdfPipelineOptions(
        do_ocr=False,              # Disable OCR for native PDFs (faster)
        do_table_structure=True,   # Enable table structure extraction
        generate_page_images=False, # Disable image generation (no need for your case)
    )

    # Create converter with custom options
    converter = DocumentConverter(
        format_options={
            InputFormat.PDF: PdfFormatOption(pipeline_options=pdf_options)
        }
    )

    # Convert the PDF to Docling model
    #result = converter.convert(pdf_path)
    doc = result.document

    # Prepare list for structured output
    output = []
    section_counter = 0
    current_section_title = None

    # Check if conversion was successful
    if "SUCCESS" in str(result.status):
        print(f"Conversion successful for {file_name}")
        
        # Iterate over each page of the converted document
        for page in doc.pages:
            page_num = page

            for block in page.blocks:

                # ---------- TEXT BLOCK ----------
                if isinstance(block, TextBlock):
                    text = block.text

                    # Detect section headings
                    if block.role and "heading" in block.role.lower():
                        section_counter += 1
                        current_section_title = text

                    output.append({
                        "type": "text",
                        "text": text,
                        "metadata": {
                            "file_name": file_name,
                            "page_number": page_num,
                            "section_number": section_counter,
                            "section_title": current_section_title,
                            "block_id": block.id,
                        }
                    })

                # ---------- TABLE BLOCK ----------
                elif isinstance(block, TableBlock):
                    output.append({
                        "type": "table",
                        "table_data": block.export_to_dataframe().to_dict(),
                        "metadata": {
                            "file_name": file_name,
                            "page_number": page_num,
                            "section_number": section_counter,
                            "section_title": current_section_title,
                            "block_id": block.id,
                        }
                    })
                else:
                    # Ignore non-text and non-table blocks (like images, graphics)
                    continue

        return output
    else:
        print(f"Conversion failed for {file_name} with status {result.status}")
        return []

# --------- USAGE ---------
pdf_path = "Documents/Attention.pdf" # Specify the path to your PDF
parsed_data = parse_pdf_with_docling(pdf_path)

# Displaying first few parsed blocks
for block in parsed_data[:5]:
    print(block)
```

```python
def clean_whitespaces(markdown_content):
    # Replace multiple spaces with a single space and remove extra newlines
    markdown_content = re.sub(r'\s+', ' ', markdown_content)  # Replace multiple spaces with one
    markdown_content = markdown_content.replace("\n", " ")    # Remove newlines
    return markdown_content.strip()  # Remove leading/trailing spaces

def remove_hyphens(markdown_content):
    markdown_content = re.sub(r'(\w+)-\n(\w+)', r'\1\2', markdown_content)
    return markdown_content

def remove_headers_footers(markdown_content, headers_footers):
    for phrase in headers_footers:
        markdown_content = markdown_content.replace(phrase, '')
    return markdown_content

def fix_broken_sentences(markdown_content):
    markdown_content = re.sub(r'(\w+)-\n(\w+)', r'\1 \2', markdown_content)  # Fix hyphenated sentences
    markdown_content = re.sub(r'(\w+)\n(\w+)', r'\1 \2', markdown_content)  # Rejoin sentences split across lines
    return markdown_content

def remove_non_text_elements(markdown_content):
    markdown_content = re.sub(r'!\[.*?\]\(.*?\)', '', markdown_content)  # Remove image tags
    return markdown_content


def preserve_headings_and_paragraphs(markdown_content):
    # Example: Ensure headings are properly formatted and paragraphs are separated
    markdown_content = re.sub(r'(#+)(.*?)\n', r'\1 \2\n', markdown_content)  # Ensure headings are correctly spaced
    markdown_content = markdown_content.replace("\n", " \n\n")  # Maintain paragraph breaks
    return markdown_content



```
